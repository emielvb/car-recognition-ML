```python
import cv2
import numpy as np
import os
import sys
```

```python
# define folders
Cars_dir = os.path.join('..', 'Data', 'Cars')
target_dir = os.path.join('..', 'Data', 'cars-rescaled')
```

With the following code we can rescale all images in the Cars datadir and save them to a separate folder. This way we have a much smaller dataset in terms of filesize, yet no actual loss of data (because our CNN only takes 256x256 input anyways).

```python
import matplotlib.pyplot as plt
```

```python
counter = 0
for brand in os.listdir(Cars_dir):
    for img in os.listdir(os.path.join(Cars_dir, brand)):
        # get full path to image
        imgpath = os.path.join(Cars_dir, brand, img)
        # read image with cv2
        cv2img = cv2.imread(imgpath)
        # rescale with cv2
        rescaled_img = cv2.resize(cv2img, (256, 256))
        # add _rescaled to name
        imgname, extension = os.path.splitext(img)
        img = '_'.join((imgname, 'rescaled')) + extension
        # save to new file location
        cv2.imwrite(os.path.join(target_dir, brand, img), rescaled_img)
        counter += 1
        print(f'Image {counter}: {img} rescaled and saved to {os.path.join(target_dir, brand)}\n')
```
