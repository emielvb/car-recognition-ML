```python
import cv2
import numpy as np
import os
import sys
```

```python
# define folders
Cars_dir = os.path.join('..', 'Data', 'Cars')
increased_dir = os.path.join('..', 'Data', 'cars-increased-size')
```

```python
# define list of allowed extensions
allowed_exts = ['jpeg', 'jpg', 'bmp', 'png']
```

```python
correct_counter = 0
for brand in os.listdir(increased_dir):
    for searchterm in os.listdir(os.path.join(increased_dir, brand)):
        counter = 0
        for img in os.listdir(os.path.join(increased_dir, brand, searchterm)):
            # get full path to image
            imgpath = os.path.join(increased_dir, brand, searchterm, img)
            # read image with cv2
            cv2img = cv2.imread(imgpath)
            # get the correct image extension
            extension = img.split('.')[-1]
            # generate systematic new image name
            imgname = '_'.join((brand, searchterm, str(counter)))
            # get image size
            imgsize = os.path.getsize(imgpath)
            # perform check on img size and extension
            if imgsize > 5000 and extension in allowed_exts:
                # save image to correct folder in Data/Cars
                targetfolder = os.path.join(Cars_dir, brand)
                imgname = '.'.join((imgname, extension))
                cv2.imwrite(os.path.join(targetfolder, imgname), cv2img)
                print(f'saved image {imgname} to {targetfolder}')

            # increase counter number
            counter += 1
```
